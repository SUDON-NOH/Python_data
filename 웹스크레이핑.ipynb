```python
# URL을 설정: 기상청 육상 중기 예보
API = 'http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp'

# stnld : 중기예보 지역코드 3자리

values = {
#   'stnld':'109' # 서울, 인천, 경기
    'stnId':'133' # 충청남도
}

# URL 요청시 필요한 param 생성
params = urllib.parse.urlencode(values)
print(params) # stnld = 133
# param을 주소뒤에 ?와 붙여 get방식으로 요청
url = API + "?" + params
print("url=", url)

# 다운로드
data = REQ.urlopen(url).read()
text = data.decode("utf-8")
print(text)
```

```python
# 서울 강남 강남역

baseurl = 'http://www.menupan.com'
murl = 'restaurant/bestrest/bestrest.asp?pt=rt&areacode=ss201'

url = baseurl + '/' + murl

response = REQ.urlopen(url)
soup = BeautifulSoup(response, 'html.parser')

rank1 = soup.select_one('.restName').string # ' . ' + Class name 하면 class로 되어 있는 곳을 찾아낸다.

urlList = soup.select_one('.rankingList .list')
urlList

listData = []
href = []
for li in urlList.select('li'):
    rank = li.select_one('.numTop, .rankNum').string 
    # select_one('A B')쉼표가 없는 경우는 A 밑에 B 이고
    # select_one('A,B') 쉼표가 있는 경우는 A or B의 의미
    listname = li.select_one('.listName').string
    listtype = li.select_one('.listType').string
    listArea = li.select_one('.listArea').string
    href.append(li.select_one('.listName a')['href']) # a는 태그를 의미
    listData.append({'랭킹':rank,
                     '상호명':listname,
                     '업종':listtype,
                     '지역':listArea})




listData_2 = []
for x in range(0, len(href) - 1):
    url = baseurl + href[x]
    response = REQ.urlopen(url)
    soup = BeautifulSoup(response, 'html.parser')
    
    main = soup.select_one('.areaBasic .restTel')
    main2 = soup.select_one('.areaBasic .restAdd')
    
    address = main2.select_one('.add1').string
    
    Tel = main.select_one('.tel1').string
    listData_2.append({'주소':address,
                       '전화번호': Tel})
print(listData_2)

# '랭킹'
# '상호명'
# '업종'
# '지역'

import pandas as pd
df = pd.DataFrame(listData)
# df
```

```python
# 대전 대덕구

baseurl = 'http://www.menupan.com'
murl = 'restaurant/bestrest/bestrest.asp?pt=rt&areacode=dj201'

url = baseurl + '/' + murl

response = REQ.urlopen(url)
soup = BeautifulSoup(response, 'html.parser')

rank1 = soup.select_one('.restName').string

urlList = soup.select_one('.rankingList .list')
urlList

listData = []
href = []
for li in urlList.select('li'):
    rank = li.select_one('.numTop, .rankNum').string 
    listname = li.select_one('.listName').string
    listtype = li.select_one('.listType').string
    listArea = li.select_one('.listArea').string
    href.append(li.select_one('.listName a')['href'])
    listData.append({'랭킹':rank,
                     '상호명':listname,
                     '업종':listtype,
                     '지역':listArea})
    

listData_2 = []
for x in range(0, len(href) - 1):
    url = baseurl + href[x]
    response = REQ.urlopen(url)
    soup = BeautifulSoup(response, 'html.parser')
    
    main = soup.select_one('.areaBasic .restTel')
    main2 = soup.select_one('.areaBasic .restAdd')
    
    address = main2.select_one('.add1').string
    
    Tel = main.select_one('.tel1').string
    listData_2.append({'주소':address,
                       '전화번호': Tel})

import pandas as pd
df = pd.DataFrame(listData)
df2 = pd.DataFrame(listData_2)

df['주소'] = df2['주소']
df['전화번호'] = df2['전화번호'] # 주소와 전화번호를 원본 데이터에 병합
df.set_index('랭킹', inplace=True) # 랭킹 컬럼을 index로 사용
addr = df2['주소'].values
df.to_excel('대전_맛집_list.xlsx') # 저장

print(addr)
df
```
