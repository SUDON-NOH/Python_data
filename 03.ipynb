# 접근성

```python
import pandas as pd
import matplotlib.pyplot as plt

import numpy as np



import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'

data = pd.read_csv('data.csv')
a = 923.58125

# accessibility (접근성)
data1 = data.replace(to_replace = '_', value = 0, regex = True)
data_b = data1.astype({'예상매출액':float})
data_p = data[(data_b['예상매출액'] < a) & (data_b['예상매출액'] > 0)]

data_accessibility = data_p[['버스정류장 거리', '지하철역 거리', '도로까지의거리', '버스정류장수', '지하철역수', '예상매출액']]

```

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler
scaler = MinMaxScaler()

data_accessibility[['버스정류장 거리', '지하철역 거리', '도로까지의거리', '예상매출액']] = \
    scaler.fit_transform(data_accessibility[['버스정류장 거리', '지하철역 거리', '도로까지의거리', '예상매출액']])


data_accessibility.corr()
```

```python
import seaborn as sns
from scipy.stats import norm

corrmat = data_accessibility.corr()

f, ax = plt.subplots(figsize = (9 , 8))
sns.heatmap(corrmat, ax = ax, cmap = 'YlGnBu', linewidths = 0.3)
```

# 인구

```python
data_pop = data_p[['거주인구세대수', '거주인구인구수', '전체 직장인구수', '남성_직장인구수', '여성_직장인구수', '유동인구합계',
                   '주재인구합계', '보행인구합계', '주간상주인구_합계', '야간상주인구_합계', '예상매출액']]


data_pop[['거주인구세대수', '거주인구인구수', '전체 직장인구수', '남성_직장인구수', '여성_직장인구수', '유동인구합계',
                   '주재인구합계', '보행인구합계', '주간상주인구_합계', '야간상주인구_합계', '예상매출액']] = \
    scaler.fit_transform(data_pop[['거주인구세대수', '거주인구인구수', '전체 직장인구수', '남성_직장인구수', '여성_직장인구수', '유동인구합계',
                   '주재인구합계', '보행인구합계', '주간상주인구_합계', '야간상주인구_합계', '예상매출액']])

corrmat = data_pop.corr()

f, ax = plt.subplots(figsize = (9 , 8))
sns.heatmap(corrmat, ax = ax, cmap = 'YlGnBu', linewidths = 0.3)
```
