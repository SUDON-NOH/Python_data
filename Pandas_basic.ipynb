#### DataFrame의 스칼라 값(원소) 접근하기: 읽기 및 수정
- (1) df.loc[행 번호, 열]  : 명시적인 index    # 열을 column name으로 접근 - 변경시 수정이 필요
- (2) df.iloc[행 번호, 열] : 함묵적인 index    # 열을 인덱스로 접근 - column name이 변경되어도 유지

```python
# df.loc[행, 열] - [] 괄호를 넣으면 여러 column에 동시에 접근 가능
print(df.loc[3, 'Country'])
df.loc[:6,'Country']
df.loc[:6, ['Country', 'CountryID']]
```

```python
# df.iloc[행, 열]

data = df.iloc[0, 0]
print(data, type(data))
                                    # <class 'str'>

data = df.iloc[2, 2]
print(data, type(data))
                                    # <class 'numpy.int64'> 

data = df.iloc[2, [0, 1, 2]]
print(data, type(data))
                                    # Country      Algeria
                                    # CountryID          3
                                    # Continent          3
                                    # Name: 2, dtype: object <class 'pandas.core.series.Series'>
                
data = df.iloc[:3, [0, 1, 2]]
print(data, type(data))
                                    #        Country  CountryID  Continent
                                    # 0  Afghanistan          1          1
                                    # 1      Albania          2          2
                                    # 2      Algeria          3          3
                                    # <class 'pandas.core.frame.DataFrame'>


df.iloc[3, 0] = 'Dorra'
print(df.iloc[:5 , 0])

df.iloc[-1, 0]
df.iloc[0, -1]
```

## 데이터 프레임과 통계

```python
sunspots.describe()   # 통계 요약
```

```python
sunspots.count()    # NaN이 아닌 값의 개수
```

```python
sunspots.mad()    # Mean Absolute Deviation 평균 절대 편차
```

```python
sunspots.std()    # Standard Deviation , 표준편차
```

```python
sunspots.median()   # 중앙값, 2분위수
```

```python
sunspots.sum()      # 합계
```

```python
sunspots.max()
```

```python
sunspots.mode()    # 최빈값을 반환한다.
```

```python
sunspots.var()    # 분산
```

```python
sunspots.skew()   # 비대칭도
```

```python
sunspots.kurt()   # 첨도를 반환한다. 첨도는 분포의 뾰족한 정도를 나타내는 지표
```

```python
# Dataframe을 불리언 방식으로 검색
# print(sunspots['Number of Observations'].means)
sunspots[sunspots > sunspots.mean()]   # 조건에 맞지 않은 값들은 전부 NaN으로 채워지고 조건에 맞는 값은 남는다.
                                       # 불필요한 결과

# 올바른 결과
sunspots[sunspots['Number of Observations'] > sunspots['Number of Observations'].mean()] # 조건에 맞는 값들을 뽑는데, NaN을 전부 제외
```
