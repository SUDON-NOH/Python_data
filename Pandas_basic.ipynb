```python
# df.loc[행, 열] - [] 괄호를 넣으면 여러 column에 동시에 접근 가능
print(df.loc[3, 'Country'])
df.loc[:6,'Country']
df.loc[:6, ['Country', 'CountryID']]
```

```python
# df.iloc[행, 열]

data = df.iloc[0, 0]
print(data, type(data))
                                    # <class 'str'>

data = df.iloc[2, 2]
print(data, type(data))
                                    # <class 'numpy.int64'> 

data = df.iloc[2, [0, 1, 2]]
print(data, type(data))
                                    # Country      Algeria
                                    # CountryID          3
                                    # Continent          3
                                    # Name: 2, dtype: object <class 'pandas.core.series.Series'>
                
data = df.iloc[:3, [0, 1, 2]]
print(data, type(data))
                                    #        Country  CountryID  Continent
                                    # 0  Afghanistan          1          1
                                    # 1      Albania          2          2
                                    # 2      Algeria          3          3
                                    # <class 'pandas.core.frame.DataFrame'>


df.iloc[3, 0] = 'Dorra'
print(df.iloc[:5 , 0])

df.iloc[-1, 0]
df.iloc[0, -1]
```

```python
# merge()

dests = pd.read_csv('dest.csv')
print(dests)
tips = pd.read_csv('tips.csv')
print(tips)

# 내부조인 : 공통조건에 만족되는 부분만 조인된다.
# 공통조건 = 'EmpNr'
pd.merge(dests, tips, how = 'inner', on = 'EmpNr')
# how = 'inner'는 공통 부분이 있는 것들만 표시
# pd.merge(dests, tips, how = 'inner' on = 'Dest') Error
# on 은 기준이 되는 열로 on 의 값을 기준으로 Join 한다.

# 외부조인 : 공통조건 관계 없이 모두 조인된다.
pd.merge(dests, tips, how = 'outer')
# how = 'outer'는 공통 부분이 없는 것들도 NaN으로 표시

```

```python
# join(self, other, on, how, lsuffix, rsuffix, sort)
dests.join(tips,lsuffix = '_Dest', rsuffix = '_Tips')
```

```python
# groupby() 함수
import numpy as np
import pandas as pd
from numpy.random import seed
from numpy.random import rand
from numpy.random import randint
seed(42)
# seed 를 고정시켜야 rand 로 돌려도 rand 의 값이 고정된다. 따라서 데이터를 분석할 때 seed를 고정시켜놓고 한다.
# 만약 고정하지 않고 돌린다면, 데이터를 분석하는 사람들 사이에서 시간의 차이가 발생할 수 있다.
# rand(7)
# np.random.seed(42)
# np.random.rand(n) # 0과 1 사이의 값을 균일분포로 난수 n개 생성
# np.random.randint(start, end + 1, size)
# randint(1, 9, 3) array([4, 7, 4])

df = pd.DataFrame({ 'Weather' : ['cold', 'hot', 'cold', 'hot', 'cold', 'hot', 'cold'],
                   'Food' : ['soup', 'soup', 'icecream', 'chocolate', 'icecream', 'icecream', 'soup'],
                   'Price' : 10 * rand(7), 'Number' : randint(1, 9, 7)})
df

```

```python
weather_group = df.groupby('Weather')
print('='*138)
print(weather_group)
print('='*138)
print(weather_group.first())
print('='*138)
print(weather_group.last())
print('='*138)
print(weather_group.describe())
print('='*138)
print(weather_group.mean())
print('='*138)
i = 0
for name, group in weather_group:
    i = i + 1
    print(i, name)
    print(group)
    print(type(group))  # <class 'pandas.core.frame.DataFrame'>
```

```python
wf_group = df.groupby(['Weather', 'Food'])
print('='*138)
print(wf_group)
print('='*138)
print(type(wf_group.groups)) # dict
print('='*138)
print(wf_group.groups)
```
