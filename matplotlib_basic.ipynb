```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('transcount.csv')
df.shape   # (81, 2)
df = df.groupby('year').agg(np.mean) # agg -> method(sum/mean 등등 가능)
df
```

```python
years = df.index.values
years
counts = df['trans_count'].values
counts

plt.subplot(2,2,1)
plt.plot(years, counts) # 올바른 정보를 알 수 없다.
plt.subplot(2,2,2)
plt.plot(years, np.log(counts)) # log를 취한 후 1970~2000년 사이의 변화가 보이게 된다.
plt.show()
```

```python
e = 2.71828182845904 # 오일러 상수, 자연로그의 밑
np.log(e)# 자연로그

a = np.log(10000)
print(a)
np.exp(a)
```

```python
poly = np.polyfit(years, np.log(counts), deg=1) # 학습
# polyfit 은 산점도에서 직선의 방정식을 구하는 함수 / deg = 함수의 차수
# polyfit(x, y, deg) : x값과 y값을 사용하여 기울기를 구함
# Least Squeares polynomial fit
# 선과 데이터의 차이가 가장 적은 선
print('Poly: ', poly)
print('Poly: ', poly[0]) # w, 기울기
print('Poly: ', poly[1]) # b, y절편
plt.semilogy(years, counts, 'o') # 산점도 # semilogy가 log 값으로 변환해줌
plt.semilogy(years, np.exp(np.polyval(poly, years))) 
# exp : 지수로 바꿈 # 위에서 log로 바꿨던 값을 다시 지수로 바꾼 후 semilogy로 로그 값으로 다시 만들어줌
plt.show()

# Hypothesis : 가설 or 예측
# f(x) = 0.3615592104293512 * 12 - 705.7831945013617
# predict = np.exp(f(x))

# 예측
print('<예측>')
predict = np.exp(np.polyval(poly, [2015,2016,2017,2019,2020]))
print('2015년:', predict[0])
print('2019년:', predict[3])
print('2020년:', predict[4])
```

#### 분산플롯(scatter plot)

```python
df = pd.read_csv('transcount.csv')
df = df.groupby('year').agg(np.mean)

gpu = pd.read_csv('gpu_transcount.csv')
gpu = gpu.groupby('year').agg(np.mean)

df = pd.merge(df, gpu, how='outer', left_index = True, right_index = True)
# 뒤에 left_index , right_index 가 없으면 공통된 부분이 없어서 merge 되지 않는다.
# df['gpu_trans_count'].mean()
# np.mean(df['gpu_trans_count'].values)  # numpy는 1개라도 NaN 이 있으면 NaN으로 결과 값을 출력

df.fillna(0)
df = df.replace(np.nan,0) # nan을 0으로 대체
years = df.index.values
print(years)
counts = df['trans_count'].values
gpu_counts = df['gpu_trans_count'].values
cnt_log = np.log(counts)

# plt.scatter(x, y, z)    x = x축 / y = y축 / z = size
# c: color, s: size, alpha = 축
plt.scatter(years, cnt_log, c = 200 * years,
            s = 20 + 200 * gpu_counts/gpu_counts.max(),
           alpha=0.5) # alpha = 투명도를 얘기한다. 작을수록 투명해짐 # 버블차트
plt.plot(years, cnt_log)
plt.grid(True)
plt.xlabel('Year')
plt.ylabel('Trans_Count')
plt.title('trans/gpu_counts per year')
plt.show()
```

#### 범례 및 주석

```python
#===========================기본데이터 그래프====================================
df = pd.read_csv('transcount.csv')
df = df.groupby('year').agg(np.mean)

gpu = pd.read_csv('gpu_transcount.csv')
gpu = gpu.groupby('year').agg(np.mean)

df = pd.merge(df, gpu, how='outer', left_index = True, right_index = True)
# 뒤에 left_index , right_index 가 없으면 공통된 부분이 없어서 merge 되지 않는다.
# df['gpu_trans_count'].mean()
# np.mean(df['gpu_trans_count'].values)  # numpy는 1개라도 NaN 이 있으면 NaN으로 결과 값을 출력

df.fillna(0)
df = df.replace(np.nan,0) # nan을 0으로 대체
years = df.index.values
counts = df['trans_count'].values
gpu_counts = df['gpu_trans_count'].values
cnt_log = np.log(counts)

poly = np.polyfit(years, cnt_log, deg = 1)
plt.plot(years, np.polyval(poly, years), label = 'Poly Fit')

#========================화살표의 위치==========================================

gpu_start = gpu.index.values.min()
# 1997 # 가장 작은 값을 찾아서 사용해서 x 값을 찾음
y_ann = np.log(df.at[gpu_start, 'trans_count']) 
# gpu_start에 대응되는 y의 값을 구함
print(y_ann)
ann_str = 'First GPU\n %d' % gpu_start # 이 값을 사용해서 annotate에 활용
print(ann_str)
plt.annotate(ann_str,xy = (gpu_start, y_ann),
            arrowprops = dict(arrowstyle = '->'),
            xytext= (-30 , +70),
            textcoords = 'offset points') 
# xy = 화살표의 위치  # ann_str 은 글자
# xytext = (x ,y) 텍스트의 위치 (조정 가능)
# offset points 는 '상대적인 값'이라는 뜻
# - x 의 값이 양수면 오른쪽 음수면 왼쪽
# - y 의 값이 양수면 위쪽 음수면 아래쪽

# plt.scatter(x, y, z)    x = x축 / y = y축 / z = size
# c: color, s: size, alpha = 축
plt.scatter(years, cnt_log, c = 200 * years,
            s = 20 + 200 * gpu_counts/gpu_counts.max(),
           alpha=0.5, label = 'Scatter Plot') # alpha = 투명도를 얘기한다. 작을수록 투명해짐 # 버블차트


#=================================범례 출력====================================

plt.legend(loc = 'upper left') # 범례를 출력
# loc ( location )으로 위치 조정 가능
# upper left or right
# lower left or right

plt.grid(True)
plt.xlabel('Year')
plt.ylabel('Trans_Count')
plt.title('trans/gpu_counts per year')
plt.show()
```

```python
#===========================기본데이터 그래프====================================
df = pd.read_csv('transcount.csv')
df = df.groupby('year').agg(np.mean)

gpu = pd.read_csv('gpu_transcount.csv')
gpu = gpu.groupby('year').agg(np.mean)

df = pd.merge(df, gpu, how='outer', left_index = True, right_index = True)
# 뒤에 left_index , right_index 가 없으면 공통된 부분이 없어서 merge 되지 않는다.
# df['gpu_trans_count'].mean()
# np.mean(df['gpu_trans_count'].values)  # numpy는 1개라도 NaN 이 있으면 NaN으로 결과 값을 출력

df.fillna(0)
df = df.replace(np.nan,0) # nan을 0으로 대체
years = df.index.values
counts = df['trans_count'].values
gpu_counts = df['gpu_trans_count'].values
cnt_log = np.log(counts)

#===========================3D 그래프 만들기====================================
print(df.shape) # 2차원 데이터
fig = plt.figure()
fig
ax = Axes3D(fig)

X = df.index.values
print(X.shape) # (37, ) 1차원 데이터
Y =  np.where(df['trans_count'].values > 0,
             np.log(df['trans_count'].values), 0)
Z = np.where(df['gpu_trans_count'].values > 0,
             np.log(df['gpu_trans_count'].values), 0)
# result = np.where(conditions, a, b)
# result is a if contions == True else b


X,Y = np.meshgrid(X, Y)
Z,_ = np.meshgrid(Z, 0) # _ 는 아무런 변수가 없다. Z만 쓰면 meshgrid의 결과가 모두 Z에 튜플로 형성된다.
# print(X.shape) # (37, 37) 1차원인 데이터를 3D 데이터를 만들기 위해 2차원으로 변경해서 생성

# Z = Z.reshape(1, Z.shape[0]) # 1차원인 데이터를 2차원으로 변경


ax.plot_surface(X, Y, Z)
ax.set_xlabel('Year')
ax.set_ylabel('Log CPU transistor Counts')
ax.set_zlabel('Log GPU transistor Counts')
ax.set_title("Moorr's Law & Transistor Counts")
plt.show()
```

#### Boxplot(상자그림)

```python
#================================기본데이터=====================================================
df = pd.read_csv('transcount.csv')
df = df.groupby('year').agg(np.mean)

gpu = pd.read_csv('gpu_transcount.csv')
gpu = gpu.groupby('year').agg(np.mean)

df = pd.merge(df, gpu, how='outer', left_index = True, right_index = True)
# 뒤에 left_index , right_index 가 없으면 공통된 부분이 없어서 merge 되지 않는다.
# df['gpu_trans_count'].mean()
# np.mean(df['gpu_trans_count'].values)  # numpy는 1개라도 NaN 이 있으면 NaN으로 결과 값을 출력

df['trans_count'] = np.log(df['trans_count'].values)
df['gpu_trans_count'] = np.log(df['gpu_trans_count'].values)

df = df.replace(np.nan,0) # nan을 0으로 대체

df.plot.box()
plt.show()
```

```python
# matplotlib 사용
plt.boxplot([df['trans_count'], df['gpu_trans_count']],
           labels = ['trans_count','gpu_trans_count'])
plt.text(1,df['trans_count'].median(),
        round(df['trans_count'].median(),2))    # 첫 df['trans_count'].median()은 위치, 
                                                # 값은 다음 df['trans_count'].median()
                                                # round의 두번째 숫자는 소숫점 자리 수
plt.text(2,df['gpu_trans_count'].median(),
        df['gpu_trans_count'].median())
plt.show()
```
